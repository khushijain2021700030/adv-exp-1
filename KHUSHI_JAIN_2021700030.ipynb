```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'], format='%d-%m-%Y')

df_merged = pd.merge(df1, df2, on='Order ID')

df_merged['Month-Year'] = df_merged['Order Date'].dt.to_period('M')

monthly_sales = df_merged.groupby('Month-Year')['Amount'].sum()

monthly_sales_growth = monthly_sales.pct_change().fillna(0) * 100

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales_growth.index.astype(str), monthly_sales_growth, marker='o', color='purple')
plt.title('Monthly Sales Growth Rate')
plt.xlabel('Month-Year')
plt.ylabel('Growth Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
df_merged = pd.merge(df1, df2, on='Order ID')

state_data = df_merged.groupby('State').agg(
    total_sales=('Amount', 'sum'),
    total_profit=('Profit', 'sum'),
    order_count=('Order ID', 'count')
).reset_index()

plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    state_data['total_sales'], 
    state_data['total_profit'], 
    s=state_data['order_count'] * 20,  
    alpha=0.6, 
    c=state_data['total_profit'], 
    cmap='coolwarm', 
    edgecolor='black'
)

plt.colorbar(scatter, label='Total Profit')


for i, state in state_data.iterrows():
    plt.annotate(
        state['State'],
        (state['total_sales'], state['total_profit']),
        fontsize=9,
        ha='right',
        va='bottom'
    )

plt.title('Order Amount vs. Profit by State')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```
